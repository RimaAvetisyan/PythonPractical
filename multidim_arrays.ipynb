```python
def list_sum(lst):
    result = 0
    for val in lst:
        result += val  # result = result + val
    return result


def matrix_sum(matrix):
    result = 0
    for row in matrix:
        result += list_sum(row)
    return result 


def max_row_matrix(matrix):
    max_mean = 0
    for row in matrix:
        max_mean += max(row)

    return max_mean / len(matrix)


def get_diagonal(matrix):
    return [matrix[i][i] for i in range(len(matrix))]


def trace(matrix):
    return sum(get_diagonal(matrix)) 


def is_sorting(lst):
    is_ascending = True
    is_descending = True

    if len(lst) <= 1:
        return 'Not sorted'

    for i in range(1, len(lst)):
        if lst[i] > lst[i - 1]:
            is_descending = False
        if lst[i] < lst[i - 1]:
            is_ascending = False
        if not is_ascending and not is_descending:
            return 'Not sorted'

    if is_ascending and is_descending:
        print('Error: getting both is ascending and is_descening. lst: %s' % (lst,))

    if is_ascending:
        return 'Ascending'
    if is_descending:
        return 'Descending'


def get_num_digits(num):
    digits = []
    while num // 10 != 0:
        digits.append(num % 10)
        num //= 10
    
    digits.append(num)

    return digits


def get_list_step_2_elems(lst, odd):
    return [lst[i] for i in range(1 if odd else 0, len(lst), 2)]


def checking(n):
    digits = get_num_digits(n)
    odd_sum = sum(get_list_step_2_elems(digits, True))
    even_sum = sum(get_list_step_2_elems(digits, False))

    return odd_sum == even_sum


def digits_sum(n):
    return sum(get_num_digits(n))


a = int(input())
print(checking(a))
```

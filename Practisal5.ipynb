```python
import random


def matrix_product(matrix_1, matrix_2):
    """
    Return matrix product result.
    If cannot product, return None.
    """
    matrix_1_row = len(matrix_1)
    matrix_1_column = len(matrix_1[0])
    matrix_2_row = len(matrix_2)
    matrix_2_column = len(matrix_2[0])

    if matrix_1_column != matrix_2_row:
        return None

    matrix_result = []
    for _ in range(matrix_1_row):
        matrix_result.append([0] * matrix_2_column)

    for i in range(matrix_1_row):
        for j in range(matrix_2_column):
            element = 0
            for k in range(matrix_1_column):
                element += matrix_1[i][k] * matrix_2[k][j]
            matrix_result[i][j] = element

    return matrix_result


def random_matrix(row, column):
    matrix = []

    for _ in range(row):
        row = []
        for _ in range(column):
            row.append(random.randint(0, 10))

        matrix.append(row)

    return matrix


def write_matrix(matrix, filename):
    """
    Write matrix to file filename.
    """
    with open(filename, 'w') as f:
        for row in matrix:
            row_str = ' '.join(map(lambda element: str(element), row))
            f.write(row_str + '\n')


def read_matrix(filename):
    """
    Read matrix from file filename.
    """
    matrix = []

    with open(filename, 'r') as f:
        lines = f.readlines()
        for row in lines:
            row_list = list(map(lambda el: int(el), row[:-1].split(' ')))
            matrix.append(row_list)

    return matrix


print(read_matrix('matrix.txt'))
```
